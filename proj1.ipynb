```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

```

```python
data = pd.read_csv("/Users/prudhveer/Desktop/dataset.csv",header=None) 
data=data.replace(to_replace="M",value=1)
data=data.replace(to_replace="B",value=0)
print(data)
```

```python
y = data.iloc[:,1].values # Assigning Output Values Column to y
print(y)
```

```python
x=data.iloc[:, 2:32] # Assigning feature columns to x
print(x)
```

```python
from sklearn.model_selection import train_test_split # Using split function to split test , train and validate in ratio 80, 10, 10 respectively
x_train_i, x_test, y_train_i, y_test = train_test_split(x, y, test_size = 0.10, random_state = 22)
x_train, x_val, y_train, y_val = train_test_split(x_train_i, y_train_i, test_size = 1/9, random_state = 1)
```

```python
sc = StandardScaler() # Using Standard Scalar to normalize the data
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
x_val = sc.fit_transform(x_val)
print(x_train)
print(x_test)
print(x_val)
```

```python

def opt(X, Y, w, b, epochs, alpha):
    
    costs=[]
    predictions=[]
    
    for i in range(epochs):
 
        number_of_training_examples = X.shape[1] #Number of training examples

    # Function for calculating Gradient Descent 
        Z = np.dot(w.T, X) + b;    
        sigmoid = 1/(1 + np.exp(- Z))
        cost= -(1/number_of_training_examples) * np.sum(Y * np.log(sigmoid) + (1-Y) * np.log(1-sigmoid)) 
    
    
        d_w = (1/number_of_training_examples)* np.dot(X, (sigmoid-Y).T)
        d_b = (1/number_of_training_examples)* np.sum(sigmoid-Y)
        # Simulataneously updating weights and bias 
        w = w - alpha * d_w
        b = b - alpha * d_b
        
        if i % 10 == 0:
            costs.append(cost) # Storing the costs after 10 iterations predict loss.
            print("cost after %i iteration : %f" % (i, cost))
        
        accuracy1 = accuracy(X, w, b)
        predictions.append(100 - np.mean(np.abs(accuracy1 - Y.T)) * 100)
        
    w = {"w":w, "b":b}
    grads = {"dw":d_w, "db":d_b}
    
    return w, grads, costs, predictions
```

```python
def accuracy(X, w, b):
    
    number_of_training_examples = X.shape[1] 
    
    y_prediction =  np.zeros((1,number_of_training_examples))
    
    w = w.reshape(X.shape[0], 1)
    
    z=(np.dot(w.T, X)+b)
    sigmoid = 1/(1 + np.exp(- z))
    

    for i in range(sigmoid.shape[1]):
        
        if(sigmoid[0,i]<0.5): #Assigning the threshold functions 
            y_prediction[0,i]=0
        else:
            y_prediction[0,i]=1           
    
    return y_prediction
```

```python
m = x_train.T.shape[0] #Number of features
w = np.zeros((m,1))
b= 0
parameters, gradients, costs, predictions = opt(x_train.T, y_train.T, w, b, epochs=600, alpha=0.03) 
weights = parameters["w"]
bias = parameters["b"]

#Drawing plots between cost and number of iterations
plt.plot(costs)
plt.title("Cost vs Number of Iterations")
plt.xlabel("Number of Iterations * 10")
plt.ylabel("Cost")
```

```python
y_pred_test = accuracy(x_test.T, weights, bias) #Fininding the predicted values for test data by using the trained model
y_pred_val = accuracy(x_val.T, weights,bias) #Fininding the predicted values for validation data by using the trained model
```

```python
#plot b/w cost and number of iterations
plt.plot(predictions)
plt.title("Predtictions vs Number of Iterations")
#plot b/w pedictions and number of iterations
plt.title("Accuracy vs Number of Iterations")
plt.xlabel("Number of Iterations * 10")
plt.ylabel("accuracy")
```

```python
from sklearn.metrics import confusion_matrix
confused_matrix = confusion_matrix(y_test, y_pred_test.T)  # Building the confusion matrix using the test data and the predicted data
print(confused_matrix)
```

```python
accuracy=((confused_matrix[0][0]+confused_matrix[1][1])/((confused_matrix[0][0]+confused_matrix[1][1])+confused_matrix[1][0]+confused_matrix[0][1]))*100
print (" Accuracy:")
print(accuracy) # TP+TN/ TP+TN+FP+FN
b=confused_matrix[1][0]
a=confused_matrix[0][0]

precision=(a/(a+b))*100
print ("precision:") #TP/TP+FP
print(precision)

recall=confused_matrix[0][0]/((confused_matrix[0][0])+(confused_matrix[0][1]))*100
print ("recall:")
print(recall) #TP/TP+FN

fmeasure=(2*recall*precision)/(recall+precision)
print ("fmeasure:")
print(fmeasure)
```
